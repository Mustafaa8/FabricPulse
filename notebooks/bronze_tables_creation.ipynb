```python
transactions = spark.read.csv("Files/bronze/transactions.csv",header=True)
customers = spark.read.csv("Files/bronze/customers.csv",header=True)
accounts = spark.read.csv("Files/bronze/accounts.csv",header=True)
instruments = spark.read.csv("Files/bronze/instruments.csv",header=True)
```

```python
transactions.write.format('delta').mode("overwrite").saveAsTable("Bronze.raw_transactions")
customers.write.format('delta').mode("overwrite").saveAsTable("Bronze.raw_customers")
accounts.write.format('delta').mode("overwrite").saveAsTable("Bronze.raw_accounts")
instruments.write.format('delta').mode("overwrite").saveAsTable("Bronze.raw_instruments")
```
